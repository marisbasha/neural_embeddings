### Evaluate the model by visualization

```python
from model import WhisperSegmenterFast
from audio_utils import SpecViewer
import pandas as pd
import librosa
segmenter = WhisperSegmenterFast( "model/whisperseg-base-bengalese-finch/final_checkpoint_ct2", device="cuda" )
spec_viewer = SpecViewer()
```

```python
audio_file = "data/datasets/bengalese-finch/test/gr41rd51__3part_SYLc_th4191_belowhits_220612_1956.16442_0.wav"
label_file = audio_file[:-4] + ".csv"
audio, sr = librosa.load( audio_file, sr = None )
label = pd.read_csv( label_file )

## by default the min_frequency parameter when calling the segment function (used to compute Log-Mel spectrogram) is 0.
## For certain species, like mouse, set the parameter min_frequency to a proper value, e.g., 35000
prediction = segmenter.segment( audio, sr = sr )
spec_viewer.visualize( audio = audio, sr = sr, prediction = prediction, label=label )
```
